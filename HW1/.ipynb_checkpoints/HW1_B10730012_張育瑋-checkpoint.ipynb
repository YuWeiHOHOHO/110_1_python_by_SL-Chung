```python
with open("ç—…æ­·500å¥å­.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    vac = dict() #dictionary
    f.seek(0) #å¾é ­è®€å–
    
    while len(f_content) > 0: #è‹¥é‚„æœ‰æ–‡å­—å°±ç¹¼çºŒè®€å–
        f_content = f.read(1)
        if f_content in vac.keys():
            vac[f_content] += 1
        else:
            vac[f_content] = 1

    #åˆªé™¤é›œé …
    to_be_deleted = '\n:/ã€‹â€¦ï¼›â€”ã€ã€‚ï¼Œ 5ï¼‰ï¼ˆ'
    key_to_be_deleted = [key for key in vac if set(key).intersection(to_be_deleted) != set({})]
    for key in key_to_be_deleted:
        vac.pop(key)
```

### å»ºç«‹å‡½å¼

```python
def N_gram(file_name, N):
    with open(file_name, "r", encoding = "UTF-8") as f:
        f_content = f.read()
        n_dic = dict() #n-gram dictionary
        n_minus1_dic = dict() #(n-1)-gram dictionary
        pr = dict()

        f.seek(0) #å¾é ­è®€å–

        #æ¶ˆé™¤é›œé …
        f_content = f_content.replace(' ','')
        f_content = f_content.replace('ã€‚','')
        f_content = f_content.replace('ï¼Œ','')
        f_content = f_content.replace('=','')
        f_content = f_content.replace('\n','')
        f_content = f_content.replace('ã€','')
        f_content = f_content.replace('5','')
        f_content = f_content.replace('0','')



        for c in range(len(f_content)-N+1): #è®€å–N-gramå‡ºç¾æ¬¡æ•¸
            if f_content[c:c+N] in n_dic.keys():
                n_dic[f_content[c:c+N]] += 1
            else:
                n_dic[f_content[c:c+N]] = 1

        for c in range(len(f_content)-(N-1)+1): #è®€å–(N-1)-gramå‡ºç¾æ¬¡æ•¸
            if f_content[c:c+N-1] in n_minus1_dic.keys():
                n_minus1_dic[f_content[c:c+N-1]] += 1
            else:
                n_minus1_dic[f_content[c:c+N-1]] = 1

        #è¨ˆç®—æ©Ÿç‡
        for c in range(len(f_content)-N): #è®€å–(N-1)-gramå‡ºç¾æ¬¡æ•¸
            x = n_dic[f_content[c:(c+N)]] / n_minus1_dic[f_content[c:(c+N-1)]]
            pr[f_content[c:(c+N)]] = x

        print("=========================================================")
        print("æ©Ÿç‡å‰åå¤§çš„è©")
        ke=list(pr.keys())
        rank_sort_key=sorted(ke,key= lambda ke:pr[ke],reverse=True)

        for i in rank_sort_key[:10]:
            print('{},{}'.format(i,pr[i]))
            
#         print("=========================================================")
#         print("N-gramè©çš„å‡ºç¾æ©Ÿç‡")
#         for value in {value for value in pr.values()}:
#             print(value)
#             print({key for key in pr.keys() if pr[key] == value})

        

        pr.clear
        
        f.close()
```

### åˆ—å‡º2-gramæ©Ÿç‡

```python
print('N-gram = 2\n')
N_gram("ç—…æ­·500å¥å­.txt", 2)
```

### åˆ—å‡º3-gramæ©Ÿç‡

```python
print('N-gram = 3\n')
N_gram("ç—…æ­·500å¥å­.txt", 3)
```

```python
print('N-gram = 4\n')
N_gram("ç—…æ­·500å¥å­.txt", 4)
```

```python
with open("ç—…æ­·500å¥å­.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    n_dic = dict() #n-gram dictionary
    n_minus1_dic = dict() #(n-1)-gram dictionary
    pr = dict()
    f.seek(0) #å¾é ­è®€å–
    N = 2

    #æ¶ˆé™¤é›œé …
    f_content = f_content.replace(' ','')
    f_content = f_content.replace('ã€‚','')
    f_content = f_content.replace('ï¼Œ','')
    f_content = f_content.replace('=','')
    f_content = f_content.replace('\n','')
    f_content = f_content.replace('ã€','')
    f_content = f_content.replace('5','')
    f_content = f_content.replace('0','')



    for c in range(len(f_content)-N+1): #è®€å–N-gramå‡ºç¾æ¬¡æ•¸
        if f_content[c:c+N] in n_dic.keys():
            n_dic[f_content[c:c+N]] += 1
        else:
            n_dic[f_content[c:c+N]] = 1

    for c in range(len(f_content)-(N-1)+1): #è®€å–(N-1)-gramå‡ºç¾æ¬¡æ•¸
        if f_content[c:c+N-1] in n_minus1_dic.keys():
            n_minus1_dic[f_content[c:c+N-1]] += 1
        else:
            n_minus1_dic[f_content[c:c+N-1]] = 1

    #è¨ˆç®—æ©Ÿç‡
    for c in range(len(f_content)-N): #è®€å–(N-1)-gramå‡ºç¾æ¬¡æ•¸
        x = n_dic[f_content[c:(c+N)]] / n_minus1_dic[f_content[c:(c+N-1)]]
        pr[f_content[c:(c+N)]] = x

    print("è©²å­—ç‚º ğ‘¥1 æ™‚ï¼Œå¯ä»¥ 100% æ–·å®šå…¶ä¸‹ä¸€å€‹å­— ğ‘¥2 å‡ºç¾çš„å­—ã€‚\n")
    print({key[0] for key in pr.keys() if pr[key] == 1.0})

    pr.clear
    n_dic.clear
    n_minus1_dic.clear
    
    f.close()
```

```python
from pypinyin import pinyin

with open("ç—…æ­·500å¥å­.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    word_to_pinyin = {} #n-gram dictionary

    f.seek(0) #å¾é ­è®€å–

    #æ¶ˆé™¤é›œé …
    f_content = f_content.replace(' ','')
    f_content = f_content.replace('ã€‚','')
    f_content = f_content.replace('ï¼Œ','')
    f_content = f_content.replace('=','')
    f_content = f_content.replace('\n','')
    f_content = f_content.replace('ã€','')
    f_content = f_content.replace('5','')
    f_content = f_content.replace('0','')

    #å»ºç«‹[æ–‡å­—:æ‹¼éŸ³]çš„dict
    for c in range(len(f_content)-1): #è®€å–N-gramå‡ºç¾æ¬¡æ•¸
        if f_content[c:c+2] not in word_to_pinyin.keys():
            word_to_pinyin[f_content[c:c+2]] = str(pinyin(f_content[c:c+2]))

    print("=========================================================")
    print("åŒéŸ³ç•°å­—è€…")
    
    for value in {value for value in word_to_pinyin.values()}:

        same_pinyin = [] #å»ºç«‹åŒéŸ³çš„list
        for key in word_to_pinyin.keys(): #è‹¥åŒéŸ³å°±åŠ å…¥list
            if str(word_to_pinyin[key]) == value:
                same_pinyin.append(key)
        
        if len(same_pinyin)>1: #å°å‡ºlistå…§å®¹è¶…éä¸€å€‹çš„åŒéŸ³è©
            print(value)
            print(same_pinyin)
    
    f.close()
```
