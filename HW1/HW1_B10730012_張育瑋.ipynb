## 第一題

### 建立函示

```python
with open("病歷500句子.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    vac = dict() #dictionary
    f.seek(0) #從頭讀取
    
    while len(f_content) > 0: #若還有文字就繼續讀取
        f_content = f.read(1)
        if f_content in vac.keys():
            vac[f_content] += 1
        else:
            vac[f_content] = 1

    #刪除雜項
    to_be_deleted = '\n:/》…；—、。， 5）（'
    key_to_be_deleted = [key for key in vac if set(key).intersection(to_be_deleted) != set({})]
    for key in key_to_be_deleted:
        vac.pop(key)
```

### 總字數

```python
print("總共用了" + str(len(vac)) + " 個不同的字\n")
```

### 每字出現次數

```python
print("每個字的出現次數")
for value in {value for value in vac.values()}:
    print(value)
    print({key for key in vac.keys() if vac[key] == value})
```

### 前十常出現的字及次數

```python
print("前十常出現的字和其出現次數")
ke=list(vac.keys())
rank_sort_key=sorted(ke,key= lambda ke:vac[ke],reverse=True)

for i in rank_sort_key[:10]:
	print('{},{}'.format(i,vac[i]))
```

### 只出現過一次的字

```python
print("只出現一次的字\n")
print({key for key in vac.keys() if vac[key] == 1})
```

## 第二題

### 建立函式

```python
def N_gram(file_name, N):
    with open(file_name, "r", encoding = "UTF-8") as f:
        f_content = f.read()
        n_dic = dict() #n-gram dictionary
        n_minus1_dic = dict() #(n-1)-gram dictionary
        pr = dict()

        f.seek(0) #從頭讀取

        #消除雜項
        f_content = f_content.replace(' ','')
        f_content = f_content.replace('。','')
        f_content = f_content.replace('，','')
        f_content = f_content.replace('=','')
        f_content = f_content.replace('\n','')
        f_content = f_content.replace('、','')
        f_content = f_content.replace('5','')
        f_content = f_content.replace('0','')



        for c in range(len(f_content)-N+1): #讀取N-gram出現次數
            if f_content[c:c+N] in n_dic.keys():
                n_dic[f_content[c:c+N]] += 1
            else:
                n_dic[f_content[c:c+N]] = 1

        for c in range(len(f_content)-(N-1)+1): #讀取(N-1)-gram出現次數
            if f_content[c:c+N-1] in n_minus1_dic.keys():
                n_minus1_dic[f_content[c:c+N-1]] += 1
            else:
                n_minus1_dic[f_content[c:c+N-1]] = 1

        #計算機率
        for c in range(len(f_content)-N): #讀取(N-1)-gram出現次數
            x = n_dic[f_content[c:(c+N)]] / n_minus1_dic[f_content[c:(c+N-1)]]
            pr[f_content[c:(c+N)]] = x

        print("=========================================================")
        print("機率前十大的詞")
        ke=list(pr.keys())
        rank_sort_key=sorted(ke,key= lambda ke:pr[ke],reverse=True)

        for i in rank_sort_key[:10]:
            print('{},{}'.format(i,pr[i]))
            
#         print("=========================================================")
#         print("N-gram詞的出現機率")
#         for value in {value for value in pr.values()}:
#             print(value)
#             print({key for key in pr.keys() if pr[key] == value})

        

        pr.clear
        
        f.close()
```

### 列出2-gram機率

```python
print('N-gram = 2\n')
N_gram("病歷500句子.txt", 2)
```

### 列出3-gram機率

```python
print('N-gram = 3\n')
N_gram("病歷500句子.txt", 3)
```

### 列出4-gram機率

```python
print('N-gram = 4\n')
N_gram("病歷500句子.txt", 4)
```

```python
with open("病歷500句子.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    n_dic = dict() #n-gram dictionary
    n_minus1_dic = dict() #(n-1)-gram dictionary
    pr = dict()
    f.seek(0) #從頭讀取
    N = 2

    #消除雜項
    f_content = f_content.replace(' ','')
    f_content = f_content.replace('。','')
    f_content = f_content.replace('，','')
    f_content = f_content.replace('=','')
    f_content = f_content.replace('\n','')
    f_content = f_content.replace('、','')
    f_content = f_content.replace('5','')
    f_content = f_content.replace('0','')



    for c in range(len(f_content)-N+1): #讀取N-gram出現次數
        if f_content[c:c+N] in n_dic.keys():
            n_dic[f_content[c:c+N]] += 1
        else:
            n_dic[f_content[c:c+N]] = 1

    for c in range(len(f_content)-(N-1)+1): #讀取(N-1)-gram出現次數
        if f_content[c:c+N-1] in n_minus1_dic.keys():
            n_minus1_dic[f_content[c:c+N-1]] += 1
        else:
            n_minus1_dic[f_content[c:c+N-1]] = 1

    #計算機率
    for c in range(len(f_content)-N): #讀取(N-1)-gram出現次數
        x = n_dic[f_content[c:(c+N)]] / n_minus1_dic[f_content[c:(c+N-1)]]
        pr[f_content[c:(c+N)]] = x

    print("該字為 𝑥1 時，可以 100% 斷定其下一個字 𝑥2 出現的字。\n")
    print({key[0] for key in pr.keys() if pr[key] == 1.0})

    pr.clear
    n_dic.clear
    n_minus1_dic.clear
    
    f.close()
```

```python
from pypinyin import pinyin
import os
os.system("cls")

with open("病歷500句子.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    pinyin_count = dict() #n-gram dictionary

    f.seek(0) #從頭讀取

    #消除雜項
    f_content = f_content.replace(' ','')
    f_content = f_content.replace('。','')
    f_content = f_content.replace('，','')
    f_content = f_content.replace('=','')
    f_content = f_content.replace('\n','')
    f_content = f_content.replace('、','')
    f_content = f_content.replace('5','')

    for c in range(len(f_content)): #讀取N-gram出現次數
        if str(pinyin(f_content[c])) in pinyin_count.keys():
            pinyin_count[str(pinyin(f_content[c]))] += 1
        else:
            pinyin_count[str(pinyin(f_content[c]))] = 1
    
    f.close()
```

## 第五題

```python
from pypinyin import pinyin

with open("病歷500句子.txt", "r", encoding = "UTF-8") as f:
    f_content = f.read()
    word_to_pinyin = {} #n-gram dictionary

    f.seek(0) #從頭讀取

    #消除雜項
    f_content = f_content.replace(' ','')
    f_content = f_content.replace('。','')
    f_content = f_content.replace('，','')
    f_content = f_content.replace('=','')
    f_content = f_content.replace('\n','')
    f_content = f_content.replace('、','')
    f_content = f_content.replace('5','')
    f_content = f_content.replace('0','')

    #建立[文字:拼音]的dict
    for c in range(len(f_content)-1): #讀取N-gram出現次數
        if f_content[c:c+2] not in word_to_pinyin.keys():
            word_to_pinyin[f_content[c:c+2]] = str(pinyin(f_content[c:c+2]))

    print("=========================================================")
    print("同音異字者")
    
    for value in {value for value in word_to_pinyin.values()}:

        same_pinyin = [] #建立同音的list
        for key in word_to_pinyin.keys(): #若同音就加入list
            if str(word_to_pinyin[key]) == value:
                same_pinyin.append(key)
        
        if len(same_pinyin)>1: #印出list內容超過一個的同音詞
            print(value)
            print(same_pinyin)
    
    f.close()
```
